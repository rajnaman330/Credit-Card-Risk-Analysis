```python
import pandas as pd

credit_df = pd.read_csv("credit_risk.csv")

credit_df.head(6)
```

```python
#We donn't need ID, so dropping it
credit_df = credit_df.drop(["ID"], axis=1)
```

```python
#renaming PAY_0 to PAY_1 for correct numbering
credit_df.rename(columns={"PAY_0":"PAY_1"}, inplace=True)
```

```python
credit_df.head()
```

```python
credit_df['EDUCATION'].value_counts()
```

```python
#Merging all higher education above 4(University) into 4 as well as merging 0 and 3 marriages 
credit_df["EDUCATION"]=credit_df["EDUCATION"].map({0:4,1:1,2:2,3:3,4:4,5:4,6:4})
credit_df["MARRIAGE"]=credit_df["MARRIAGE"].map({0:3,1:1,2:2,3:3})
```

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
X = credit_df.drop(['default.payment.next.month'], axis = 1)
X = scaling.fit_transform(X)
```

```python
Y = credit_df['default.payment.next.month']
```

```python
## Hyperparameter optimization using RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
```

```python
# Importing the datetime module to work with date and time
from datetime import datetime

# Function to track the time elapsed since a specified start time
def timer(start_time=None):
    # If no start time is provided, set the current time as the start time
    if not start_time:
        start_time = datetime.now()
        return start_time
    # If a start time is provided, calculate the elapsed time from the start time
    elif start_time:
        # Calculate the total elapsed time in seconds
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        # Convert total seconds to hours and remaining seconds
        tmin, tsec = divmod(temp_sec, 60)
        # Print the formatted elapsed time
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
classifier=xgboost.XGBClassifier()
```

```python
random_search=RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring='roc_auc',n_jobs=-1,cv=5,verbose=3)
```

```python
from datetime import datetime

start_time = timer(None) # timing starts from this point for "start_time" variable
random_search.fit(X,Y)
timer(start_time) # timing ends here for "start_time" variable
```
